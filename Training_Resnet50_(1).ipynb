```python

def extract_frames(video_path, output_dir, frame_limit=10):
    """
    Extract frames from a video and save them to the specified output directory.

    Args:
        video_path (str): Path to the video file.
        output_dir (str): Directory to save the extracted frames.
        frame_limit (int): Number of frames to extract from each video.

    Returns:
        list: Paths to the extracted frames.
    """
    os.makedirs(output_dir, exist_ok=True)
    cap = cv2.VideoCapture(video_path)
    frame_paths = []
    count = 0

    while cap.isOpened() and count < frame_limit:
        ret, frame = cap.read()
        if not ret:
            break

        frame_filename = os.path.join(output_dir, f"{os.path.basename(video_path)}_frame_{count}.jpg")
        cv2.imwrite(frame_filename, frame)
        frame_paths.append(frame_filename)
        count += 1

    cap.release()
    return frame_paths

def preprocess_data(imgs_paths, labels, frame_output_dir="/content/extracted_frames"):
    """
    Process image and video paths, extracting frames from videos and flattening the list.

    Args:
        imgs_paths (list): List of image or video paths.
        labels (list): List of corresponding labels.
        frame_output_dir (str): Directory to store extracted frames.

    Returns:
        tuple: (processed_img_paths, processed_labels)
    """
    processed_imgs = []
    processed_labels = []

    for img_path, label in zip(imgs_paths, labels):
        if img_path.lower().endswith(('.mp4', '.avi', '.mov', '.mkv')):  # Check for video formats
            video_frames = extract_frames(img_path, frame_output_dir)
            processed_imgs.extend(video_frames)  # Add all extracted frames
            processed_labels.extend([label] * len(video_frames))  # Assign the same label to each frame
        else:
            processed_imgs.append(img_path)  # Keep image path as is
            processed_labels.append(label)

    return processed_imgs, processed_labels

# Apply preprocessing
train_imgs_processed, train_labels_processed = preprocess_data(train_imgs, train_labels)
val_imgs_processed, val_labels_processed = preprocess_data(val_imgs, val_labels)
test_imgs_processed, test_labels_processed = preprocess_data(test_imgs, test_labels)

print("Training set:", len(train_imgs_processed), "images")
print("Validation set:", len(val_imgs_processed), "images")
print("Test set:", len(test_imgs_processed), "images")
```

```python
class CustomDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        """
        Args:
            image_paths (list): List of paths to image files.
            labels (list): List of labels corresponding to the image paths.
            transform (callable, optional): Optional transform to be applied on an image.
        """
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        label = self.labels[idx]

        # Open image
        image = Image.open(img_path).convert("RGB")

        # Apply transformations if specified
        if self.transform:
            image = self.transform(image)

        # Convert label to tensor
        label = torch.tensor(label, dtype=torch.long)

        return image, label
```

```python
# Define transformations for train, validation, and test sets
data_transforms = {
    'train': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
    'validation': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
    'test': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
}

# Create datasets
train_dataset = CustomDataset(train_imgs_processed, train_labels_processed, transform=data_transforms['train'])
val_dataset = CustomDataset(val_imgs_processed, val_labels_processed, transform=data_transforms['validation'])
test_dataset = CustomDataset(test_imgs_processed, test_labels_processed, transform=data_transforms['test'])

# Create data loaders
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=4)
val_loader = torch.utils.data.DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=4)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=32, shuffle=False, num_workers=4)

# Combine data loaders into a dictionary
dataloaders = {
    'train': train_loader,
    'validation': val_loader,
    'test': test_loader
}

# Dataset sizes for loss normalization
dataset_sizes = {
    'train': len(train_dataset),
    'validation': len(val_dataset),
    'test': len(test_dataset)
}
```

```python
# Load ResNet50 model
model = models.resnet50(pretrained=True)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Modify the fully connected layer to match the number of classes
model.fc = nn.Linear(model.fc.in_features, 2)  # Assuming binary classification (e.g., violence vs non-violence)
model = model.to(device)

# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)

```

```python

def train_and_validate_model(model, criterion, optimizer, dataloaders, dataset_sizes, num_epochs=25):
    for epoch in range(num_epochs):
        print(f"Epoch {epoch+1}/{num_epochs}")
        print("-" * 10)

        # Training and validation phases
        for phase in ['train', 'validation']:
            if phase == 'train':
                model.train()
            else:
                model.eval()

            running_loss = 0.0
            running_corrects = 0

            for inputs, labels in dataloaders[phase]:
                inputs = inputs.to(device)
                labels = labels.to(device)
                optimizer.zero_grad()

                with torch.set_grad_enabled(phase == 'train'):
                    outputs = model(inputs)
                    _, preds = torch.max(outputs, 1)
                    loss = criterion(outputs, labels)

                    if phase == 'train':
                        loss.backward()
                        optimizer.step()

                running_loss += loss.item() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)

            epoch_loss = running_loss / dataset_sizes[phase]
            epoch_acc = running_corrects.double() / dataset_sizes[phase]

            print(f"{phase} Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}")

    return model
```

```python
# Train and validate the model
model = train_and_validate_model(model, criterion, optimizer, dataloaders, dataset_sizes, num_epochs=25)
```

```python
# Save the trained model
torch.save(model.state_dict(), "resnet50_multiclass_detection_trained_final.pth")
print("Model saved!")
```

```python
def evaluate_model(model, dataloader, device):
    """
    Evaluate the model and calculate accuracy, precision, recall, confusion matrix,
    and classification report.

    Args:
        model: Trained PyTorch model.
        dataloader: DataLoader for the test/validation set.
        device: Device to run the model (CPU or GPU).

    Returns:
        A dictionary containing accuracy, precision, recall, confusion matrix,
        and classification report.
    """
    model.eval()  # Set model to evaluation mode
    all_preds = []
    all_labels = []

    with torch.no_grad():  # Disable gradient computation
        for inputs, labels in dataloader:
            inputs, labels = inputs.to(device), labels.to(device)

            # Forward pass
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)  # Get predicted class

            all_preds.extend(preds.cpu().numpy())  # Store predictions
            all_labels.extend(labels.cpu().numpy())  # Store true labels

    # Calculate metrics
    accuracy = accuracy_score(all_labels, all_preds)
    precision = precision_score(all_labels, all_preds, average='weighted')  # Weighted for imbalanced datasets
    recall = recall_score(all_labels, all_preds, average='weighted')
    cm = confusion_matrix(all_labels, all_preds)
    report = classification_report(all_labels, all_preds)

    return {
        "accuracy": accuracy,
        "precision": precision,
        "recall": recall,
        "confusion_matrix": cm,
        "classification_report": report
    }
```

```python
# Evaluate the model
evaluation_results = evaluate_model(model, dataloaders['test'], device)

# Print the results
print(f"Accuracy: {evaluation_results['accuracy']:.4f}")
print(f"Precision: {evaluation_results['precision']:.4f}")
print(f"Recall: {evaluation_results['recall']:.4f}")
print("Confusion Matrix:")
print(evaluation_results['confusion_matrix'])
print("\nClassification Report:")
print(evaluation_results['classification_report'])
```
